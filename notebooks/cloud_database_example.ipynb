# Anta Database Cloud Example

This notebook demonstrates how to use the Anta Database with cloud-based zarr data stored on S3.

## Setup

First, let's import the necessary packages:

```python
import sys
sys.path.insert(0, '..')  # Add project root to path

from anta_database.cloud_database import CloudDatabase
import matplotlib.pyplot as plt
```

## Connect to Cloud Database

We'll connect to the public Anta Database on S3:

```python
# Initialize CloudDatabase
db = CloudDatabase(s3_bucket='anta-database')

# Test the connection
if db.test_connection():
    print('✅ Successfully connected to S3 database')
else:
    print('⚠️ Could not test S3 connection, but continuing...')
```

## Query for Flight Data

Let's query for a specific flight line (DC_LDC_DIVIDE):

```python
# Query for the flight line
results = db.query(flight_id='DC_LDC_DIVIDE', retain_query=True)

# Show query results
print(f'Found flight lines: {results["flight_id"]}')
print(f'Available variables: {results["var"]}')
print(f'Available ages: {results["age"]}')
```

## Plot Transect Data

Now we'll generate a 1D transect plot using the zarr data from S3:

```python
# Generate the transect plot
# This uses zarr data directly from S3 - no local files needed!
db.plot.transect_1D(
    title='DC_LDC_DIVIDE Transect - Cloud Zarr Data',
    marker_size=0.5,
    scale_factor=1.0
)
```

## Additional Plot Types

The CloudDatabase supports all the same plotting methods as the local Database:

```python
# Plot by dataset (shows all flight lines in the dataset)
# Note: This may take a moment as it downloads data from S3
db.plot.dataset(
    downsampling_factor=10,  # Reduce data points for faster plotting
    marker_size=0.3,
    title='Dataset Overview - Cloud Zarr Data'
)
```

## Clean Up

Always remember to close the database connection:

```python
# Close the database connection
db.close()
print('✅ Database connection closed')
```

## Summary

This notebook demonstrated:

1. **Cloud Database Connection**: Connecting to the Anta Database on S3
2. **Querying**: Finding specific flight lines in the database
3. **Cloud Plotting**: Generating plots directly from zarr data on S3
4. **Seamless Integration**: Using the same API as local databases

The CloudDatabase automatically handles:
- S3 authentication (public or private buckets)
- Zarr data access from cloud storage
- Efficient data streaming (only downloads what's needed)
- All plotting functionality without local file storage

This makes it easy to work with large datasets without downloading everything locally!
